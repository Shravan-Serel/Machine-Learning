```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fifa = pd.read_csv("fifa_data.csv")
```

### Histogram

```python
bins = np.arange(40,110,10)

plt.hist(fifa.Overall, bins = bins)

plt.xticks(bins)

plt.ylabel("Number of players")
plt.xlabel("Overall")

plt.show()
```

### Piecharts

```python
left = fifa['Preferred Foot'].value_counts().get('Left',0)
#returns the count of players having preferred foot as left and returns 0 if not found
right = fifa['Preferred Foot'].value_counts().get('Right',0)

labels = ['Left','Right']
colors = ['#abcdef','#aabbcc']

plt.pie([left,right],labels = labels, colors = colors, autopct = '%.2f %%')

plt.title("Foot peference of FIFA players")

plt.show()
```
