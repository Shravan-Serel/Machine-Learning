### Filtering Data

```python
new_df = df.loc[((df['Type 1'] == 'Fire') | (df['Type 1'] == 'Water') | (df['Type 1'] == 'Grass')) & (df['Legendary'] == True)]
new_df
```

```python
new_df.reset_index(drop = True, inplace = True)
new_df
```

```python
new_df.to_csv('Filtered.csv')
```
